```
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
from matplotlib import rcParams

import seaborn as sns

%matplotlib inline
rcParams['figure.figsize'] = 10, 8
sns.set_style('whitegrid')
```

```
caminho = 'https://gist.githubusercontent.com/seankross/a412dfbd88b3db70b74b/raw/5f23f993cd87c283ce766e7ac6b329ee7cc2e1d1/mtcars.csv'
carros = pd.read_csv(caminho)
carros.columns = ['nomes','mpg', 'cyl',	'disp',	'hp',	'drat','wt','qsec',	'vs','am','gear','carb']
carros.index = carros.nomes
mpg = carros['mpg']

plt.hist(mpg)
plt.show()
```

```
sns.distplot(mpg)
```

```
carros.plot(kind ='scatter',x='hp',y='mpg',c=['darkgray'],s=150)
```

```
sns.regplot(x='hp',y='mpg',data=carros,scatter=True)
```

```
sns.pairplot(carros)
```

```
colunas =['mpg','disp','hp','wt']
novo_df = pd.DataFrame((carros[colunas].values),columns=colunas)
carros_target = carros['am'].values

novo_df['grupo'] = pd.Series(carros_target,dtype='category')
```

```
sns.pairplot(novo_df,hue='grupo',palette='hls')
```

```
carros.boxplot(column='mpg',by='am')
carros.boxplot(column='wt',by='am')
```

```
sns.boxplot(x='am',y='mpg',data=carros,palette='hls')
```
